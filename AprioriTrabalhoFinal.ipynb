```python
# -----------------------------

# 2?? Transformar em matriz One-Hot

# -----------------------------

te = TransactionEncoder()

df_bool = te.fit(transacoes).transform(transacoes)

df = pd.DataFrame(df_bool, columns=te.columns_)



print("\n?? Matriz One-Hot (itens x transações):")

display(df.head())
```

```python
# ============================
# 3?? Rodar Apriori com suporte mínimo de 15%
# ============================

frequent_items = apriori(df, min_support=0.15, use_colnames=True)
print("\n=== Itens frequentes (min_support = 0.15) ===")
display(frequent_items)
```

```python
# ============================
# Regras de associação
# ============================

regras = association_rules(frequent_items, metric="confidence", min_threshold=0.6)

# Ordenar pelas melhores regras (lift)
regras = regras.sort_values(by="lift", ascending=False)

print("\n=== Regras de Associação (confiança: 60%) ===")
display(regras)
```

```python
# -----------------------------

# 6?? Mostrar regras mais fortes

# -----------------------------

print("\n?? Regras mais fortes (ordenadas por confiança):")

for _, row in regras.iterrows():

    ant = ', '.join(list(row['antecedents']))

    cons = ', '.join(list(row['consequents']))

    print(f"{ant} ? {cons} | suporte={row['support']:.2f}, confiança={row['confidence']:.2f}, lift={row['lift']:.2f}")
```
